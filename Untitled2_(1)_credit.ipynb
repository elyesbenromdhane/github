**V** **cramer**


````python


def cramers_v(confusion_matrix):
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    r_corr = r - ((r-1)**2)/(n-1)
    k_corr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2corr / min((k_corr-1), (r_corr-1)))

# Select the categorical columns and target variable
categorical_columns = ['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'Credit_History', 'Property_Area']
target_variable = 'Loan_Status'

# Calculate Cramer's V for each categorical variable against the target variable
cramer_v_values = {}
for col in categorical_columns:
    confusion_matrix = pd.crosstab(loan_data[col], loan_data[target_variable])
    cramer_v = cramers_v(confusion_matrix)
    cramer_v_values[col] = cramer_v

# Display the results
for col, value in cramer_v_values.items():
    print(f"V de Cramer entre '{col}' et '{target_variable}': {value:.3f}")
````

#**Analyse** **Anova**

````python
from scipy.stats import f_oneway

# Assuming you have your data in a pandas DataFrame called df
loan_status = loan_data['Loan_Status']
LoanAmount = loan_data['LoanAmount']
ApplicantIncome = loan_data['ApplicantIncome']
CoapplicantIncome = loan_data['CoapplicantIncome']

# Perform separate one-way ANOVA tests for each variable
result_loan_amount = f_oneway(LoanAmount[loan_status == 1], LoanAmount[loan_status == 0])
result_applicant_income = f_oneway(ApplicantIncome[loan_status == 1], ApplicantIncome[loan_status == 0])
result_coapplicant_income = f_oneway(CoapplicantIncome[loan_status == 1], CoapplicantIncome[loan_status == 0])

print("LoanAmount ANOVA:")
print("F-statistic:", result_loan_amount.statistic)
print("p-value:", result_loan_amount.pvalue)
print("")

print("ApplicantIncome ANOVA:")
print("F-statistic:", result_applicant_income.statistic)
print("p-value:", result_applicant_income.pvalue)
print("")

print("CoapplicantIncome ANOVA:")
print("F-statistic:", result_coapplicant_income.statistic)
print("p-value:", result_coapplicant_income.pvalue)
````

#**Courbe ROC**

````python




# Prédire les probabilités de classe positive
probabilities = log_model.predict_proba(x_test)[:, 1]

# Calculer la courbe ROC
fpr, tpr, thresholds = roc_curve(y_test, probabilities)
roc_auc = auc(fpr, tpr)

# Trouver le point d'inflexion optimal
optimal_idx = np.argmax(tpr - fpr)
optimal_threshold = thresholds[optimal_idx]

# Utiliser le point d'inflexion optimal pour prédire les classes positives et négatives
predicted_classes = (probabilities >= optimal_threshold).astype(int)

# Afficher la courbe ROC et le point d'inflexion optimal
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='Courbe ROC ')
plt.plot(fpr[optimal_idx], tpr[optimal_idx], marker='o', markersize=8, label='Point d\'inflexion optimal', color='red')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Taux de faux positifs')
plt.ylabel('Taux de vrais positifs')
plt.title('Courbe ROC')
plt.legend(loc='lower right')
plt.show()

# Utiliser le point d'inflexion optimal pour prédire les classes positives et négatives dans l'ensemble de test
predicted_classes = (probabilities >= optimal_threshold).astype(int)

# Vous pouvez ensuite évaluer les performances du modèle en utilisant d'autres métriques, telles que la précision, le rappel, etc.
````
